```python
from transformers import BertTokenizer
tz = BertTokenizer.from_pretrained("bert-base-cased")
file_content = open("actsa-sentiment_Classification-Telegu.csv", encoding="utf8").read()
#import pandas as pd
#tz.tokenize(myfile.txt)
#te=pd.read_csv("myfile.txt")
#file = open("myfile.txt")
tz.tokenize(file_content) 
```

```python
pt=set(tz.tokenize(file_content))
```

```python
pt
```

```python
kw=[]
for item in pt:
    kw.append({item:len(item)})
```

```python
kw
```

```python
kww=[]
for item in pt:
    kww.append(len(item))
```

```python
kww
```

```python
#kww1=[]
kww.sort()
print(kww)
```

```python
kww
```

```python
from collections import Counter
import pandas
import matplotlib.pyplot as plt

#data = ['a','a','b','c','c']
count = Counter(kww)
df = pandas.DataFrame.from_dict(count,orient='index')

df.plot(kind='bar',width=0.5,color='g',figsize =(8,5),alpha = 0.7, fontsize = 15,legend=None)
plt.xlabel("token’s  length   ",fontsize=15)
plt.ylabel("frequency",fontsize=15)
plt.title('Task1_Result_Telgu dataset',fontsize=15)
```
